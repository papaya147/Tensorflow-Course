```python
import tensorflow as tf
```

```python
# random tensors
random_1 = tf.random.Generator.from_seed(42)
random_1 = random_1.normal(shape = (3, 2, 3))
random_1
```

```python
# tensor shuffling
shuffle_random_1 = tf.random.shuffle(random_1)
shuffle_random_1
```

```python
# tensor of ones
tf.ones((4,6))
```

```python
# tensor of zeros
tf.zeros((5,2))
```

```python
# convert numpy array to tensor
import numpy as np
array_1 = np.arange(1, 25, dtype=np.int32)
A = tf.constant(array_1, shape=(2,3,4))
A
```

```python
# rank 4 tensor
rank_4 = tf.zeros((2, 3, 4, 5))
rank_4
```

```python
# tensor attributes
rank_4.shape, rank_4.ndim, tf.size(rank_4)
```

```python
# indexing
rank_4[:2, :2, :2, :2]
```

```python
rank_4[:1, :1, :1, :]
```

```python
rank_4[:1, :1, :, :1]
```

```python
rank_2 = tf.constant(((10, 7), (3, 4)))
rank_2
```

```python
rank_2[:, -1]
```

```python
# adding dim
rank_3 = rank_2[..., tf.newaxis]
rank_3
```

```python
# also adding dim
tf.expand_dims(rank_2, axis=-1) # expanding on last axis
```

```python
# operations
tensor = tf.constant(((10, 7), (3, 4)))
tensor
```

```python
tensor + 10
```

```python
tensor * 20
```

```python
tensor - 50
```

```python
tf.multiply(tensor, 20) # tf functions faster than base op
```

```python
# matrix multiplication
tf.matmul(tensor, tensor)
```

```python
t1 = tf.constant(((1, 2, 5),
                  (7, 2, 1),
                  (3, 3, 3)))
t2 = tf.constant(((3, 5),
                  (6, 7),
                  (1, 8)))
t1, t2
```

```python
tf.matmul(t1, t2)
```

```python
tf.tensordot(t1, t2, axes=1)
```

```python
# change dtype of tensor
A = tf.constant(((1.9, 8.3),
                (4.5, 6.9)))
tf.cast(A, dtype=tf.float16)
```

```python
# tensor aggregation
B = tf.constant(np.random.randint(0, 100, size=50))
B
```

```python
tf.reduce_min(B)
```

```python
tf.reduce_max(B)
```

```python
tf.reduce_mean(B)
```

```python
tf.reduce_sum(B)
```

```python
tf.math.reduce_variance(tf.cast(B, dtype=tf.float32))
```

```python
tf.math.reduce_std(tf.cast(B, dtype=tf.float32))
```

```python
tf.argmax(B)
```

```python
tf.argmin(B)
```

```python
# squeezing
tf.random.set_seed(42)
C = tf.constant(tf.random.uniform(shape=[50]), shape=(1, 1, 1, 1, 50))
C
```

```python
tf.squeeze(C)
```

```python
# one-hot encoding
some_list = [0, 1, 2, 3]

tf.one_hot(some_list, depth=len(some_list))
```

```python
# squaring
D = tf.reshape(tf.range(1, 11), shape=(2, 5))
tf.square(D)
```

```python
# square rooting
tf.sqrt(tf.cast(D, dtype=tf.float32))
```

```python
# log
tf.math.log(tf.cast(D, dtype=tf.float32))
```
